# Logistic Regression

```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

# Target = Survived (0 or 1)

```python
df =df[['survived','pclass','sex','age','sibsp','parch','embarked']]
```

```python
df.head()
```

```python
df=df.dropna() #handle missing values 
```

```python
df_encoded=pd.get_dummies(df,columns = ['pclass','sex','embarked'],drop_first=True)
```

```python
df_encoded
```

```python
X= df_encoded.drop('survived',axis=1)
Y= df_encoded['survived']
```

```python
"""
TrainTestSplit
we use 80% of data for training and 20% of our datafor testing
if the model is good for us
"""
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=52)
```

```python
# Scalling 
# StandardScaler 
from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler() #StandardScaler scaler = new StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.fit_transform(X_test)
```

```python
X
```

```python
from sklearn.linear_model import LogisticRegression 
```

```python
model = LogisticRegression(max_iter=1000)
```

```python
model.fit(X_train_scaled,Y_train)
```

```python
model.coef_[0]
```

```python
dir(model)
```

```python
X_test
```

```python
X_test_scaled[0]
```

```python
model.predict(X_test_scaled)
```

```python
Y_test
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
Y_pred=model.predict(X_test_scaled)
```

```python
print('Accuracy: ', accuracy_score(Y_test,Y_pred)*100)
```

```python
confusion_matrix(Y_test,Y_pred)
```

```python
print(classification_report(Y_test,Y_pred))
```
