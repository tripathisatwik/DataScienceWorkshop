### Missing Values Handling 

```python
df[~df['deck'].isnull()]
```

```python
df.isnull().sum()
```

```python
df.isnull().mean()
```

```python
# Removing Columns 
df = df.drop(columns=['deck'])
df.head()
```

```python
#Fill Missing 
df['age'].fillna(0)
```

```python
# Filling with median values 
median_value = df['age'].median()
df['age']=df['age'].fillna(median_value)
```

```python
df.isna().sum()
```

```python
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])
```

```python
df['embark_towb'] = df['embark_town'].fillna(df['embark_town'].mode()[0])
```

```python
df[df.duplicated()]
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
df.columns
```

```python
corr_df=df[['survived', 'pclass','age', 'sibsp', 'parch', 'fare','alone']].corr()
sns.heatmap(corr_df,annot=True,cmap='coolwarm')
```
